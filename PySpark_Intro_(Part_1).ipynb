## Question 3: Return the License Plates containing the consecutive numbers `93`

For example `'932PAI'` should be included, but `'9FP3AI'` should not be included.

```python
# Assuming you have an RDD named license_plates_rdd containing the license plates

# Define a function to check for consecutive numbers
def has_consecutive_93(license_plate):
    # Convert the license plate to a string
    license_plate_str = str(license_plate)
    # Iterate over the characters in the license plate string
    for i in range(len(license_plate_str) - 1):
        # Check if the current character and the next character form the consecutive numbers 93
        if license_plate_str[i:i+2] == '93':
            return True
    # Return False if no consecutive numbers 93 were found
    return False

# Apply the filter transformation to the RDD using the custom function
consecutive_93_rdd = license_plates_rdd.filter(has_consecutive_93)

# Print the license plates that contain consecutive numbers 93
for license_plate in consecutive_93_rdd.collect():
    print(license_plate)
```

## Question 4: Extract the first value of each license plate and count the number of occurences for each value

For example, `'932PAI'` will be counted towards the `9` group

```python
# write your answer here
# Assuming you have an RDD named license_plates_rdd containing the license plates

# Extract the first character of each license plate
first_chars_rdd = license_plates_rdd.map(lambda plate: plate[0])

# Count the number of occurrences for each value
char_counts = first_chars_rdd.countByValue()

# Print the number of occurrences for each value
for char, count in sorted(char_counts.items()):
    print(f"Character {char}: {count} occurrences")

```
